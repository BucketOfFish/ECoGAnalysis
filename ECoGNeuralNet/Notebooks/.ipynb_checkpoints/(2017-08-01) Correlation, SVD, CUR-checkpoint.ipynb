First we load and sort the data.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import h5py
import numpy as np
import os
```

```python
# load data
data = h5py.File("../../../Data/EC2_blocks_1_8_9_15_76_89_105_CV_HG_align_window_-0.5_to_0.79_file_nobaseline.h5")
X_data = data['Xhigh gamma'][:]
y_data = data['y'][:] # phoneme label
```

```python
# preprocessing
X_data = np.divide(X_data, np.amax(X_data)) # normalize ECoG signal
```

```python
# sort by phonemes
sort_order = y_data.argsort() # sort by phonemes
X_data = X_data[sort_order]
y_data = y_data[sort_order]
```

```python
# calculate averages for each phoneme
X_averages = []
for phoneme in range(max(y_data)+1):
    indices = [y_data == phoneme]
    X_averages.append(np.mean(X_data[indices], axis=0))
```

Next we look at the SVD, PCA, and CUR decompositions of the first event, just to see what it looks like. Note that PCA is simply a subset of the results from SVD.

```python
import pymf
import seaborn as sb
```

```python
#for i in range(10):
#    pymf.SVD(X_data[i])
```

```python
SVD = pymf.SVD(X_data[0])
SVD.factorize()
```

```python
sb.heatmap(SVD.U);
plt.show()
sb.heatmap(SVD.S);
plt.show()
sb.heatmap(SVD.V);
plt.show()
```

```python
PCA = pymf.PCA(X_data[0], num_bases=40)
PCA.factorize()
```

```python
sb.heatmap(PCA.W);
plt.show()
sb.heatmap(PCA.H);
plt.show()
```

```python
CUR = pymf.CUR(X_data[0])
CUR.factorize()
```

```python
sb.heatmap(CUR.U);
plt.show()
sb.heatmap(CUR.S);
plt.show()
sb.heatmap(CUR.V);
plt.show()
```
